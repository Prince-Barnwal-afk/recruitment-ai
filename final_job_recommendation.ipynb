```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import pickle

data = pd.read_csv("resume_dataset_with_salary.csv")

def clean(col):
  return col.str.lower().str.split(";").apply(lambda x:[i.strip() for i in x])

required_skills = clean(data["skills_required"])
job_titles = data["job_title"]

# using the pretrained model
mlb = pickle.load(open("mlb.pkl", "rb"))

X = mlb.transform(required_skills)

X_train, X_test, Y_train, Y_test = train_test_split(
    X, job_titles, test_size=0.2, random_state=42
)

model = RandomForestClassifier()
model.fit(X_train, Y_train)

#important
# skills_required → job_title this will help which job required which skills
pickle.dump(model, open("job_model.pkl", "wb"))

y_prediction = model.predict(X_train)
y_pred = model.predict(X_test)

print("Accuracy on Training data",accuracy_score(Y_train,y_prediction)*100)
print("Accuracy on Testing data",accuracy_score(Y_test,y_pred)*100)
```
